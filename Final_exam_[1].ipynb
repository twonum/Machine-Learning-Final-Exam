## E 2)

If you think about the task of spam classification more thoroughly, do you think that the classification accuracy or misclassification error is a good error metric of how good an email classifier is? What are potential pitfalls? (Hint: think about false positives [non-spam email classified as spam] and false negatives [spam email classified as non-spam]).

When evaluating a classifier, such as an email spam classifier, simply relying on classification accuracy or misclassification error can be misleading. Here's why:

1) Classification Accuracy
Definition: Classification accuracy is the ratio of correctly predicted instances to the total instances.
Pitfall: While accuracy is straightforward, it doesn't account for the distribution of classes. In a situation where spam emails are rare compared to non-spam (which is often the case), a classifier might achieve high accuracy by simply predicting "non-spam" for most emails. For instance, if 95% of emails are non-spam, a classifier that always predicts "non-spam" would have 95% accuracy but would be completely useless for spam detection.
2) Misclassification Error
Definition: Misclassification error is the complement of accuracy (1 - accuracy), representing the proportion of incorrect predictions.
Pitfall: Similar to accuracy, this metric doesn't differentiate between the types of errors (false positives and false negatives). It treats all errors as equally important, which may not be the case in real-world applications like spam classification.
3) False Positives (FP) and False Negatives (FN)
False Positives (FP): A non-spam email classified as spam.

Impact: This can be quite problematic because important emails might be missed by the user if they are wrongly filtered as spam. This can lead to loss of important information or opportunities.
False Negatives (FN): A spam email classified as non-spam.

Impact: This is also undesirable because spam emails could clutter the inbox, potentially leading to security risks (e.g., phishing).
4) Why Accuracy or Error Rate May Not Be Sufficient
Imbalanced Classes: As mentioned, if the number of spam emails is much smaller than non-spam, accuracy might not reflect the model's true performance in detecting spam.
Different Cost of Errors: The cost of a false positive (important email in spam) might be different from a false negative (spam in inbox), but accuracy treats them equally.
Better Metrics for Spam Classification
Precision: The ratio of correctly predicted spam emails to all emails predicted as spam (TP / (TP + FP)). High precision means that when the classifier predicts spam, it is usually correct.
Recall (Sensitivity): The ratio of correctly predicted spam emails to all actual spam emails (TP / (TP + FN)). High recall means the classifier is good at catching most spam emails.
F1 Score: The harmonic mean of precision and recall. This balances the two metrics and is useful when you need to account for both false positives and false negatives.

Write a for loop to evaluate different *k* nn models with k=1 to k=14. In particular, fit the `KNeighborsClassifier` on the training subset, then evaluate it on the training subset, validation subset, and test subset. Report the respective classification error or accuracy.

````python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd

# Define preprocessing function
def preprocess_data(X):
    if isinstance(X, pd.DataFrame):
        for col in X.select_dtypes(include=['object']).columns:
            le = LabelEncoder()
            X[col] = le.fit_transform(X[col])
    elif isinstance(X, np.ndarray):
        # If X is already a numpy array, no preprocessing is needed
        return X
    return X

# Ensure that all inputs are in the correct format
def preprocess_labels(y):
    if isinstance(y, pd.Series):
        le = LabelEncoder()
        y = le.fit_transform(y)
    elif isinstance(y, np.ndarray):
        if np.issubdtype(y.dtype, np.object):
            le = LabelEncoder()
            y = le.fit_transform(y)
    return y

# Preprocess data
X_train_sub = preprocess_data(X_train_sub)
X_val = preprocess_data(X_val)
X_test = preprocess_data(X_test)

# Encode labels
y_train_sub = preprocess_labels(y_train_sub)
y_val = preprocess_labels(y_val)
y_test = preprocess_labels(y_test)

# Initialize lists to store accuracy for each k
train_accuracy = []
val_accuracy = []
test_accuracy = []

# Loop through k values from 1 to 14
for k in range(1, 15):
    # Initialize and fit the KNN model
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_sub, y_train_sub)
    
    # Predict on each subset
    y_train_pred = knn.predict(X_train_sub)
    y_val_pred = knn.predict(X_val)
    y_test_pred = knn.predict(X_test)
    
    # Calculate accuracy for each subset
    train_acc = accuracy_score(y_train_sub, y_train_pred)
    val_acc = accuracy_score(y_val, y_val_pred)
    test_acc = accuracy_score(y_test, y_test_pred)
    
    # Append to lists
    train_accuracy.append(train_acc)
    val_accuracy.append(val_acc)
    test_accuracy.append(test_acc)
    
    # Print results for current k
    print(f'k={k}: Train={train_acc:.2f}, Val={val_acc:.2f}, Test={test_acc:.2f}')
````

## E 20)

Consider the following code cell, where I implemented *k*-nearest neighbor classification algorithm following the the scikit-learn API

````python
import numpy as np


class KNNClassifier(object):
    def __init__(self, k, dist_fn=None):
        self.k = k
        if dist_fn is None:
            self.dist_fn = self._euclidean_dist

    def _euclidean_dist(self, a, b):
        dist = 0.
        for ele_i, ele_j in zip(a, b):
            dist += ((ele_i - ele_j)**2)
        dist = dist**0.5
        return dist

    def _find_nearest(self, x):
        dist_idx_pairs = []
        for j in range(self.dataset_.shape[0]):
            d = self.dist_fn(x, self.dataset_[j])
            dist_idx_pairs.append((d, j))

        sorted_dist_idx_pairs = sorted(dist_idx_pairs)

        return sorted_dist_idx_pairs

    def fit(self, X, y):
        self.dataset_ = X.copy()
        self.labels_ = y.copy()
        self.possible_labels_ = np.unique(y)

    def predict(self, X):
        predictions = np.zeros(X.shape[0], dtype=int)
        for i in range(X.shape[0]):
            k_nearest = self._find_nearest(X[i])[:self.k]
            indices = [entry[1] for entry in k_nearest]
            k_labels = self.labels_[indices]
            counts = np.bincount(k_labels,
                                 minlength=self.possible_labels_.shape[0])
            pred_label = np.argmax(counts)
            predictions[i] = pred_label
        return predictions
````

A more efficient way to select the *k* smallest values from an array is to use a priority queue, for example, implemented using a heap data structure. A convenient `nsmallest` function that does exactly that is available from Python's standard library:

````python
from heapq import nsmallest

nsmallest(3, c)
````

In the code cell below, use the `%timeit` magic command to benchmark the `nsmallest` function:

````python
# ENTER YOUR CODE HERE
%timeit nsmallest(3, c)
````

Summarize your findings in 1-3 sentences.

Using nsmallest is faster than using sorted when finding a few smallest elements because it only performs the necessary operations to identify those elements, avoiding the overhead of fully sorting the entire array.
